```python
Moscow_ano1 = Moscow[:12]
Moscow_ano2 = Moscow[12:24]
Moscow_ano3 = Moscow[24:36]
Moscow_ano4 = Moscow[36:48]
Moscow_ano5 = Moscow[48:60]
Moscow_ano6 = Moscow[60:72]
Moscow_ano7 = Moscow[72:84]
```

```python
#Análise sazonalidade Moscow

anos_Moscow = [Moscow_ano1, Moscow_ano2, Moscow_ano3, Moscow_ano4, Moscow_ano5, Moscow_ano6, Moscow_ano7]

for i, anos in enumerate(anos_Moscow, start = 1):
    plt.plot(
        np.arange(1,13,1), 
        anos,
        label = f'Ano {i}' )

plt.legend()
plt.xlabel('Meses')
plt.ylabel('Valor')
plt.title('Análise de sazonalidade - Moscow')
plt.show()
```

```python
# Verifica se dois arrays são iguais

np.array_equal(Moscow_ano5, Moscow_ano7)
```

```python
# Verifica se arrays são próximos (dentro de um intervalo)

np.allclose(Moscow_ano5, Moscow_ano7, 10)
```
