CÁLCULO DO COEFINCIENTE ANGULAR



- a = Coeficiente angular
- n = número de elementos
- Y = feature eixo y
- X = feature eixo x

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
a = (n*np.sum(X*Y) - np.sum(X) * np.sum(Y))/((n*np.sum(X**2)) - (np.sum(X)**2))
a
```

CÁLCULO DO COEFICIENTE LINEAR




```python
b = np.mean(Y) - a * np.mean(X)
b
```

```python
y = a*X+b
```

```python
np.linalg.norm(Moscow - y)
```
