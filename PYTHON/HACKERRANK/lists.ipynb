Initialize your list and read in the value of  followed by  lines of commands where each command will be of the  types listed above. Iterate through each command in order and perform the corresponding operation on your list.

The first line contains an integer "n", denoting the number of commands.

Considere uma lista (list = []). Você pode executar os seguintes comandos:

1. Inserir "e": Inserir um inteiro "e" na posição "i"
2. Printar: Printar a lista
3. Remover "e": Remover a primeira occorrência "e"
4. Appendar "e": Inserir um inteiro "e" no final da lista
5. Ordenar: Ordenar a lista
6. POP: Identificar o ultimo elemento da lista
7. Reverse: Inverter a lista

Inicialize a sua lista e leia o valor de "n" seguido de "n" linhas de comandos onde cada comando será um dos 7 tipos listados acima. Itere através de cada comando na ordem e realize as operações correspondentes na sua lista.

Exemplo

N=4
append1
append2
insert1_3
print

- append1: Appenda 1 à lista, arr = [1]
- append2: Appenda 2 à lista, arr = [1, 2]
- append3: Insere 3 no index 1, arr = [1,3,2]
- print: Printa a lista

Output:
[1, 3, 2]

Input Format

A primeira linha contém um inteiro "n", denotando o número de comandos.
Cada linha "i" de "n" linhas subsequentes contém um de cada comando descritos acima.

Restrição

Os elementos adicionados na lista devem ser inteiros

Output Format

Para cada comando do tipo print, imprima a lista em uma nova linha



```python
n = int(input('Digite a quantidade de comandos desejados: '))

list = []

for i in range(n):
    print('1. Inserir "e": Inserir um inteiro "e" na posição "i"')
    print('2. Printar: Printar a lista')
    print('3. Remover "e": Remover a primeira occorrência "e"')
    print('4. Appendar "e": Inserir um inteiro "e" no final da lista')
    print('5. Ordenar: Ordenar a lista')
    print('6. POP: Retira o ultimo elemento da lista')
    print('7. Reverse: Inverter a lista')

    comando = int(input('Digite o número refernte a um dos comandos acima: '))

    match comando:
        case 1:
            num = int(input('Digite o número que deseja inserir na lista: '))
            list.insert(i, num)
        case 2:
            print(list)
        case 3:
            num = int(input('Digite o número que deseja remover da lista: '))
            list.remove(num)
        case 4:
            num = int(input('Digite o número que deseja appendar na lista: '))
            list.append(num)
        case 5:
            list.sort()
        case 6:
            list.pop(-1)
        case 7:
            list.reverse()
```
